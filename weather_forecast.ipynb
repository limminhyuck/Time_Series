# 데이터셋
https://www.kaggle.com/datasets/smid80/weatherww2

```python
weather_station_location = pd.read_csv("./Weather Station Locations.csv")
weather = pd.read_csv("./Summary of Weather.csv")
```

```python
weather_station_location.head()
```

```python
weather_station_location = weather_station_location.loc[:,["WBAN", "NAME", "STATE/COUNTRY ID", "Latitude", "Longitude"]]
```

```python
weather_station_location.head()
```

```python
weather.head()
```

```python
weather = weather.loc[:,["STA", "Date", "MeanTemp"]]
weather.head()
```

```python
weather_station_id = weather_station_location[weather_station_location.NAME == "BINDUKURI"].WBAN
weather_day = weather[weather.STA == int(weather_station_id)]
weather_day["Date"] = pd.to_datetime(weather_day["Date"])
weather_day
# 1943년 5월 11일부터 1945년 5월 31일까지 일단위 평균 온도
```

```python
# 시계열 그래프
plt.figure(figsize=(22, 8))
plt.plot(weather_day.Date, weather_day.MeanTemp)
plt.title("Mean Temp of Bindukuri Area")
plt.xlabel("Date")
plt.ylabel("Mean Temp")
plt.show()
```

```python
# 시계열 분해법
timeSeries = weather_day.loc[:, ["Date", "MeanTemp"]]
timeSeries.index = timeSeries.Date
ts = timeSeries.drop("Date", axis=1)
ts
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(ts['MeanTemp'], model='additive', period=7)
# period는 분기별일 때 4, 월별 일때 12, 주별 패턴이 있는 일별 데이터는 7

fig = plt.figure()
fig = result.plot()
fig.set_size_inches(20, 20)
```

```python
# 데이터가 패턴을 보이므로 정상성이 의심됨
import statsmodels.api as sm

fig = plt.figure(figsize=(20, 8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts, lags=20, ax=ax1)
```

```python
# 천천히 값이 작아짐, 즉 ACF값이 천천히 감소하는 것은 정상성을 만족하지 않음을 의미
# ADF(Augmented Dickey-Fuller test)검정

from statsmodels.tsa.stattools import adfuller
result = adfuller(ts)
print("ADF : %f" % result[0])
print('p-value: %f' %result[1])
print('Critical Values: ')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
# p-value가 0.05를 넘기때문에 정상성을 만족하지 못함
# 1차 차분
ts_diff = ts - ts.shift()
plt.figure(figsize=(22, 8))
plt.plot(ts_diff)
plt.title("Diff")
plt.xlabel('Date')
plt.ylabel('Diff Mean Temp')
plt.show()
```

```python
# 일정한 패턴이 없고 정상성 만족하는듯 보임
# 다시 검정
result = adfuller(ts_diff[1:])
print("ADF : %f" % result[0])
print('p-value: %f' %result[1])
print('Critical Values: ')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
# p-value가 0.05보다 작으므로 1차 차분 데이터는 정상성 만족
# 정상성 만족하는 차분 데이터로 ACF와 PACF를 그래프 그리기
# ARIMA 모형 p, q 결정

fig = plt.figure(figsize=(22, 8))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_acf(ts_diff[1:], lags=20, ax=ax1)
fig = sm.graphics.tsa.plot_pacf(ts_diff[1:], lags=20, ax=ax2)
```

```python
# ACF와 PACF 모두 금방 0에 수렴, 두번쨰 lag이후 0에 수렴한다.
# 즉 ARIMA(2,1,2)을 base model

from statsmodels.tsa.arima.model import ARIMA
from datetime import datetime

# fit model
model = ARIMA(ts, order=(2,1,2))
model_fit = model.fit()

# predict
start_index = datetime(1944, 6, 25)
end_index = datetime(1945, 5, 31)
forecast = model_fit.predict(start_index, end=end_index, typ='levels')
# 이때 tpy='levels'로 차분이 들어간 모델일 경우 linear로 할 경우 차분한 결과가 나옴

plt.figure(figsize=(22,8))
plt.plot(weather_day.Date, weather_day.MeanTemp, label = "original")
plt.plot(forecast, label="predicted")
plt.title('Time Series pred')
plt.xlabel('Date')
plt.ylabel('Mean Temp')
plt.legend()
plt.show()
```

```python
# 잔차 분석 : 어떠한 패턴이나 특성이 나타나서는 안됨
residual = np.array(weather_day[weather_day.Date>=start_index].MeanTemp) - np.array(forecast)

plt.figure(figsize=(22,8))
plt.plot(weather_day.Date[weather_day.Date>=start_index], residual)
plt.xlabel('Date')
plt.ylabel('residual')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(residual, lags=20, ax=ax1)
```

```python
result = adfuller(residual)
print("ADF : %f" % result[0])
print('p-value: %f' %result[1])
print('Critical Values: ')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
# ACF 그래프가 빠르게 0으로 수렴하고, p-value값도 0.05보다 작음 
# 성능판단

from sklearn import metrics

def scoring(y_true, y_pred):
    r2 = round(metrics.r2_score(y_true, y_pred) * 100, 3)
    corr = round(np.corrcoef(y_true, y_pred)[0, 1], 3)
    mape = round(metrics.mean_absolute_percentage_error(y_true, y_pred)*100, 3)
    rmse = round(metrics.mean_squared_error(y_true, y_pred, squared=False), 3)
    
    df = pd.DataFrame({
        'R2' : r2,
        'Corr' : corr,
        'RMSE' : rmse,
        'MAPE' : mape
    }, index=[0])
    
    return df
```

```python
scoring(np.array(weather_day[weather_day.Date>=start_index].MeanTemp), np.array(forecast))
```
